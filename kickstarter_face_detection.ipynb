OpenCV face detection variable

```python
# use this block of code when the data set has already been partly analyzed and now is being continued to analyze 
csv_path = "/content/drive/MyDrive/ML_in_applied_settings/dataset/opencv_face_detection.csv"
photos_df = pd.read_csv(csv_path)
photos_df.head()


# initialization when first analyzing the data set, set is_face to -1
#photos_df = df[['id', 'photo_url']].copy()
#photos_df['is_face'] = -1
#photos_df.head()
```

```python
backends = [
  'opencv', 
  'ssd', 
  'dlib', 
  'mtcnn', 
  'retinaface', 
  'mediapipe'
]


for i in range(0, len(photos_df['id'])):
  try:
    if photos_df['is_face'][i] != -1: # skip rows that have already been analyzed using continue
      continue
    photos_df['is_face'][i] = 1 if DeepFace.extract_faces(photos_df['photo_url'][i], 
          detector_backend = backends[0], enforce_detection=False)[0]['confidence'] != 0 else 0 # extract result using the photo_url

    # save intermediate results

    if i%500 == 0:
      filename = 'opencv_face_detection.csv'
      photos_df.to_csv('/content/drive/MyDrive/ML_in_applied_settings/dataset/' + filename, index = False)
      print(i)
  except:
    # catch erroneous links
    photos_df['is_face'][i] = -2

filename = 'opencv_face_detection.csv'
photos_df.to_csv('/content/drive/MyDrive/ML_in_applied_settings/dataset/' + filename, index = False)
```
